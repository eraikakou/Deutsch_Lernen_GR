# Speaking


1. **Τι κάνεις?** = Wie geht's dir?
    - **Αυτή την στιγμή αρκετά άσχημα** = Im Moment ziemlich schlecht.
    - **έτσι και έτσι** = So la la.
    - **Λοιπόν έτσι και έτσι** = Na ja, es geht so.
    - **Ευχαριστώ, επίσης καλά** = Danke, auch gut. 
    - **Εγώ δυστυχώς δεν είμαι καθόλου καλά.** = Mir geht's leider gar nicht gut!
    - **Ευχαριστώ, πολύ καλά** = Danke, bestens.

1. **Τι κάνεις? => Είμαι ...** = Wie geht es dir? Mir geht es sehr gut/nicht so gut/...
1. **Are you interested?** = Hast du Interesse?
